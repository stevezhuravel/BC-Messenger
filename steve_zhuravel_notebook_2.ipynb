```
from urllib.request import urlopen
from bs4 import BeautifulSoup
## opens the url of the website
html = urlopen("http://www.brooklyn.cuny.edu/web/academics/schools/naturalsciences/departments/computers/news.php")

# bsObj = BeautifulSoup(html)
soup = BeautifulSoup(html, "html.parser")
## print the content of the page
print(soup)
## prints it out 
```

```
# soup_page=soup(xml_page,"xml")
news_list=soup.findAll("div", "fermata_item")

# print
lst = []

for i in news_list:
  title = i.find("a").getText()
  ## find the date
  date = i.find('span').getText()
  lst.append(title + " " + date)
 ## returned the list in reverse order 
list(reversed(lst))
## print the list out in reverse order from oldest to newest.
```
